```python
ax = sns.countplot(x='listed_in(type)',data = df,palette ='bright',linewidth=1.5,edgecolor='black')
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Type of Restuarant',size=15)
plt.title('Count of Different Restuarants',size=18,weight='bold')
plt.show()
```

Majority of Restuarants fall into dining category

```python
gp = df.groupby('listed_in(type)',as_index = False).agg({"votes":'sum'})

plt.plot(gp['listed_in(type)'],gp['votes'], c ="green",marker='o')
plt.xlabel("Type of Restuarant",size=15)
plt.ylabel("Votes",size=15)
plt.title('Type of Restaurant and Votes',size=18,weight='bold')
plt.show()
```

Dining restuarants are more preferred by restuarants as it have more votes

```python
ax=sns.countplot(x='online_order',data = df,palette='rocket',linewidth=1.5,edgecolor = 'black')
for container in ax.containers:
    ax.bar_label(container)
plt.title('Count of Orders',size=18,weight='bold')
plt.show()
```

```python
plt.hist(x='rate',data = df , bins= 10,histtype='step',color='green')
plt.title('Rating Distribution',size=18,weight='bold')
plt.show()
```

Majority of restaurants recieved ratings from 3.5 to 4

```python
plt.figure(figsize=(12,6))
cost_two = df['approx_cost(for two people)']
ax = sns.countplot(x=cost_two,palette ='tab10',linewidth = 1.5,edgecolor='black' )
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Aprroximate Cost for Two People',size=15)
plt.title('Approximate Cost for Two People',size=18,weight='bold')
```

Majority of couples prefer restaurants with an approx cost of 300 rupees

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data = df,palette='rocket')
plt.title('Orders - Rating',size=18,weight='bold')
```

Offline orders recieved lower ratings in comparison to online orders

```python
grouped = df.groupby(['listed_in(type)', 'online_order']).size().unstack(fill_value=0)
sns.heatmap(grouped,annot=True,cmap='YlGnBu',fmt='d')
plt.title("Heatmap",size=18,weight='bold')
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

Dining Restuarants primarly accept offline orders , where as cafes accept online orders.
This states that customers prefer to place orders in person at dining restuarants , but prefer online ordering at cafes

```python
gp = df.groupby(['listed_in(type)','book_table']).size().unstack()
colors = sns.color_palette("pastel")
gp.plot(kind='bar',stacked=True,color = colors)
plt.xlabel('Type of Restaurant')
plt.ylabel('Count')
plt.title('Table Booking Availability by Restaurant Type',size=15,weight='bold')
plt.xticks(rotation=45) 
plt.show()
```

Majority of dining restaurants do not offer table booking, while cafÃ©s and buffets have a smaller number of restaurants, with a few offering table booking.
